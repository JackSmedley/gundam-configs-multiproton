```python
CorrMat = np.array([[0.0169, 0.0455, -0.22035, 0.214461],
                   [0.0455, 1., -2.245, 0.9909],
                   [-0.22035, -2.245, 6.25, -4.62375],
                   [0.214461, 0.9909, -4.62375, 7.29]])

for i in range(CovMat.shape[0]):
    for j in range(CovMat.shape[1]):
        CorrMat[i,j] = CovMat[i,j]/np.sqrt(CovMat[i,i])/np.sqrt(CovMat[j,j])
        print('ZExpCorr[%d][%d] = %f;'%(i, j, CorrMat[i,j]))
```
